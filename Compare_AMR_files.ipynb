```python
import pandas as pd
from google.colab import files
import io

# Upload files
resfinder_file = files.upload()

# Load the files
# Check the uploaded file name
print("Uploaded files:", resfinder_file.keys())

# Use the correct file name from the printed keys
filename = list(resfinder_file.keys())[0]  # Automatically use the first uploaded file
df_res = pd.read_csv(io.StringIO(resfinder_file[filename].decode('utf-8')))

# Display the first few rows
print(df_res.head())


pw_amr_file = files.upload()

# Load the files
# Check the uploaded file name
print("Uploaded files:", pw_amr_file.keys())

# Use the correct file name from the printed keys
filename = list(pw_amr_file.keys())[0]  # Automatically use the first uploaded file
df_pw = pd.read_csv(io.StringIO(pw_amr_file[filename].decode('utf-8')))

# Display the first few rows
print(df_pw.head())


# Clean Resfinder gene names
resfinder_genes = {gene.replace('.match', '') for gene in df_res.columns}
resfinder_genes2 = [w.replace('_', '') for w in resfinder_genes]
print (resfinder_genes2)
# Get PW AMR gene names
pw_amr_genes = set(df_pw .columns)

# Find substrings
substr_in_resfinder = {gene for gene in pw_amr_genes if any(gene in res_gene for res_gene in resfinder_genes2)}

print("PW AMR genes that are substrings of Resfinder genes:", substr_in_resfinder)
```
